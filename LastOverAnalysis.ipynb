```python
df_id_inning = df[['id', 'inning', 'over']].copy()
```

```python
df_id_inning
```

```python
df_max_over = df_id_inning.groupby(['id', 'inning']).max()
```

```python
output_list = os.listdir('./Output/')
print(output_list)
if 'MaxFinalOver.csv' in output_list:
    print("File Exists")
    os.remove('./Output/MaxFinalOver.csv')
    df_max_over.to_csv('./Output/MaxFinalOver.csv')
else:
    print("File Not Exists")
    df_max_over.to_csv('./Output/MaxFinalOver.csv')
```

```python
df_final_over = df.merge(df_max_over, how='inner', on=['id', 'inning', 'over'])
```

```python
# df_final_over
output_list = os.listdir('./Output/')
print(output_list)
if 'df_final_over.csv' in output_list:
    print("File Exists")
    os.remove('./Output/df_final_over.csv')
    df_final_over.to_csv('./Output/df_final_over.csv')
else:
    print("File Not Exists")
    df_final_over.to_csv('./Output/df_final_over.csv')
```

```python
df_match_details = pd.read_csv('./Inputs/IPL/ipl-complete-dataset-20082020/IPL Matches 2008-2020.csv')
```

```python
df_match_details
```

```python
final_df = df_final_over.merge(df_match_details, how='inner', on=['id'])
```

```python
# final_df
# df_final_over
output_list = os.listdir('./Output/')
print(output_list)
if 'final_df.csv' in output_list:
    print("File Exists")
    os.remove('./Output/final_df.csv')
    final_df.to_csv('./Output/final_df.csv')
else:
    print("File Not Exists")
    final_df.to_csv('./Output/final_df.csv')
```

## Best Batsman by strike rate in last over (min. 200 runs scored)

```python
df_last_over_run = final_df[['batsman', 'batsman_runs']]
```

```python
df1 = df_last_over_run.groupby(['batsman'], as_index=False).sum(['batsman_runs']).sort_values(['batsman_runs'], ascending=False)
df2 = df_last_over_run.groupby(['batsman'], as_index=False).count().sort_values(['batsman_runs'], ascending=False)

df_final = df1.merge(df2, how='inner', on=['batsman'])
df_final
df_final['strike_rate'] = round((df_final['batsman_runs_x'] / df_final['batsman_runs_y']) * 100, 2)
batsman_strike_rate = df_final[df_final['batsman_runs_x'] >= 100].sort_values(['strike_rate'], ascending=False)
batsman_strike_rate.to_csv('./Output/batsman_strike_rate.csv')
batsman_strike_rate
```

## Best Bowler by Economy (min. 60 balls bowled)

```python
df_last_over_run_conceeded = final_df[['bowler', 'total_runs']]
```

```python
# df1 = df_last_over_run_conceeded.groupby(['bowler']).sum().sort_values(['total_runs'], ascending=False)
# df2 = df_last_over_run_conceeded.groupby(['bowler']).count().sort_values('[total_runs]', ascending=False)
# df_final = df1.merge(df2, how='inner', on=['bowler'])
# df_final
df1 = df_last_over_run_conceeded.groupby(['bowler'], as_index=False).sum().sort_values(['total_runs'], ascending=False)
df2 = df_last_over_run_conceeded.groupby(['bowler'], as_index=False).count().sort_values(['total_runs'], ascending=False)
df_final = df1.merge(df2, how='inner', on=['bowler'])
df_final['economy'] = round(df_final['total_runs_x'] / (df_final['total_runs_y'] / 6), 2)
bowler_economy = df_final[df_final['total_runs_y'] >= 60].sort_values(['economy'])
bowler_economy
```

## Best Strike rate for bowlers

```python
df_last_over_details = final_df[['bowler', 'dismissal_kind']]
```

```python
wicketType = ['caught', 'bowled', 'lbw', 'stumped', 'caught and bowled']
bowler_wicket = df_last_over_details[df_last_over_details['dismissal_kind'].isin(wicketType)].groupby(['bowler']).count()
bowler_strike_rate = bowler_economy.merge(bowler_wicket, how='inner', on=['bowler'])
bowler_strike_rate['strike_rate'] = round(bowler_strike_rate['total_runs_y'] / bowler_strike_rate['dismissal_kind'], 2)
bowler_strike_rate
```

### Comparison between Batsman and Bowler according to the rankings in the respective year

```python
final_df.columns
```

```python
df1 = final_df[['batsman', 'bowler', 'batsman_runs', 'date']].copy()
df1['date'] = pd.to_datetime(df1['date'])
df1['Year'] = df1['date'].dt.year
df1['Month'] = df1['date'].dt.month_name()
print(df1['Month'].unique())
print(df1)
```

```python
df2 = df1.groupby(['batsman','bowler', 'Year', 'Month'], as_index=False).sum(['batsman_runs'])
df3 = df2.merge(batsman_strike_rate, how='inner', on=['batsman'])
df3 = df3.drop(columns=['batsman_runs_x', 'batsman_runs_y', 'strike_rate'])
df3
```

```python
df_IPL_ESPN_Mapping = pd.read_csv('./MappingFiles/IPL_ESPN_Mapping.csv', header=None)
new_column_names = ['IPL', 'ESPN']
df_IPL_ESPN_Mapping.columns = new_column_names
df_IPL_ESPN_Mapping
```

```python
df_Ratings = df3.merge(df_IPL_ESPN_Mapping, how='inner', left_on=['batsman'], right_on=['IPL'])
df_Ratings = df_Ratings.rename(columns={'ESPN' : 'ESPN Batsman'})
df_Ratings = df_Ratings.drop(columns=['batsman', 'IPL'])
df_Ratings = df_Ratings.merge(df_IPL_ESPN_Mapping, how='inner', left_on=['bowler'], right_on=['IPL'])
df_Ratings = df_Ratings.rename(columns={'ESPN': 'ESPN Bowler'})
df_Ratings = df_Ratings.drop(columns=['bowler', 'IPL'])
df_Ratings
```

```python
batsman_Ratings_files = os.listdir('./Ratings/T20/Batsman/')
df_temp = []
for file in batsman_Ratings_files:
    df_temp1 = pd.read_csv('./Ratings/T20/Batsman/' + file)
    df_temp.append(df_temp1)
df_Batsman_Ratings = pd.concat(df_temp, ignore_index=False)
df_Batsman_Ratings = df_Batsman_Ratings[df_Batsman_Ratings['Rank'] <= 50]
# df_Batsman_Ratings

batsman_Ratings_mapping = pd.read_csv('./MappingFiles/Batsman_Ratings_ESPN.csv')
batsman_Ratings_mapping = batsman_Ratings_mapping.rename(columns={'Unnamed: 0': 'Ranking', '0': 'ESPN'})
# batsman_Ratings_mapping

Batsman_Ratings = df_Batsman_Ratings.merge(batsman_Ratings_mapping, how='inner', left_on=['Name'], right_on=['Ranking'])
# df_Batsman_Ratings = df_Batsman_Ratings.drop(columns=['Name', 'Ranking'])
Batsman_Ratings = Batsman_Ratings.drop(columns=['Name', 'Ranking'])
Batsman_Ratings = Batsman_Ratings.rename(columns={'ESPN': 'ESPN Batsman'})
Batsman_Ratings
```

```python
bowler_Ratings_files = os.listdir('./Ratings/T20/Bowler/')
df_temp = []
for file in bowler_Ratings_files:
    df_temp1 = pd.read_csv('./Ratings/T20/Bowler/' + file)
    df_temp.append(df_temp1)
df_Bowler_Ratings = pd.concat(df_temp, ignore_index=False)
df_Bowler_Ratings = df_Bowler_Ratings[df_Bowler_Ratings['Rank'] <= 50]
# df_Bowler_Ratings

bowler_Ratings_mapping = pd.read_csv('./MappingFiles/Bowler_Ratings_ESPN.csv')
bowler_Ratings_mapping = bowler_Ratings_mapping.rename(columns={'Unnamed: 0': 'Ranking', '0': 'ESPN'})
# bowler_Ratings_mapping

Bowler_Ratings = df_Bowler_Ratings.merge(bowler_Ratings_mapping, how='inner', left_on=['Name'], right_on=['Ranking'])
# df_Bowler_Ratings = df_Bowler_Ratings.drop(columns=['Name', 'Ranking'])
Bowler_Ratings = Bowler_Ratings.drop(columns=['Name', 'Ranking'])
Bowler_Ratings = Bowler_Ratings.rename(columns={'ESPN': 'ESPN Bowler'})
Bowler_Ratings
```
