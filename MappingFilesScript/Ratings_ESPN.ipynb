```python
import pandas as pd
import os
import requests
from bs4 import BeautifulSoup
import time
```

```python
batsman_ratings_path = "./../Ratings/T20/Batsman/"
bowler_ratings_path = './../Ratings/T20/Bowler/'
```

```python
files = os.listdir(batsman_ratings_path)
files_bowler = os.listdir(bowler_ratings_path)
```

```python
batsman_df = []
bowler_df = []
```

```python
# df = pd.read_csv('./../Ratings/T20/Batsman/20080131.csv')
# df.loc[(df['Rank'] <= 50)]
```

```python
for file in files:
    # print(batsman_ratings_path + file)
    df = pd.read_csv(batsman_ratings_path + file)
    filter_df = df.loc[(df['Rank'] <= 100)]
    batsman_df.append(filter_df)
```

```python
for file in files_bowler:
    # bowler_df.append(pd.read_csv(bowler_ratings_path + file))
    df = pd.read_csv(bowler_ratings_path + file)
    filter_df = df.loc[(df['Rank'] <= 100)]
    bowler_df.append(filter_df)
```

```python
batsman_df = pd.concat(batsman_df, ignore_index=True)
bowler_df = pd.concat(bowler_df, ignore_index=True)
```

```python
batsman_array = batsman_df['Name'].unique()
bowler_array = bowler_df['Name'].unique()
```

```python
# method to get the full name of player from the espn crickinfo website

def getFullName(name):
    time.sleep(3)
    temp = name + " cricket player"
    googleLink = 'https://www.google.com/search'
    response = requests.get(googleLink, params={'q': temp}).text
    print(requests.get(googleLink, params={'q': temp}).url)
    soup = BeautifulSoup(response, features='html.parser').find_all('a')
    # soup = BeautifulSoup(response, features='html.parser')
    # print(soup)
    # name = soup.find_all("h3", class_ = "name link-cta")
    try:
        # print(soup) 
        # allLinks = [link.get('href') for link in soup]
        # all_links = []
        # print(soup)
        for url_link in soup:
            # print(url_link)
            temp_url = url_link.get("href")
            # print(temp_url)
            if "espncricinfo" in temp_url and "img" not in temp_url and "image" not in temp_url:
                temp_url = temp_url.replace("/url?q=","").split('&')[0]
                # print(temp_url)
                break
        responsePlayer = requests.get(temp_url).text
        soup = BeautifulSoup(responsePlayer, features='html.parser')
        # print(soup)
        fullName = soup.find_all("span", class_ = "ds-text-title-s ds-font-bold ds-text-typo")[0].find("p").text
        return fullName
    except:
        return "NA"
```

```python
def player_espn_mapping(array):
    dict = {}
    for player in array:
        player = str(player)
        if "," not in player and "(sub)" not in player:
            try:
                # print(str(player))
                fullName = getFullName(str(player))
                print(fullName)
                dict[player] = fullName
            except:
                print("NA")
        # df_processed.replace(player, getFullName(player), inplace=True)
    # print(getFullName("RT Ponting"))
    # df_processed.replace('RT Ponting', 'Rickey Ponting')
    # df_processed
    return dict
```
